## Introduction
We are considering a ball of mass $m$ rolling in a double-well potential (whatever that is) of $V(x) = x^4/4 - x^2/2$. This potential is unaptly called the "sombrero potential" despite its likeness toward stingrays.

```python
x = np.linspace(-1.5, 1.5, 100)
v = (x**4)/4 - (x**2)/2

plot = plt.figure(figsize=(12,8))
cartesian = plt.axes()
cartesian.plot(x, v, color = 'b')

cartesian.set(xlabel = "x", ylabel = "V(X)", title = "Sombrero?")
cartesian.set(xlim = (-1.5,1.5))
plt.show()
```

Taking physics things into acount and also remembering that the ball must follow Newton's Second Law, we obtain the equation:
$$m\ddot{x} = f_{\text{hat}}(x) + f_{\text{drag}}(\dot{x}) + f_{\text{drive}}(t) = x - x^3 - \nu \dot{x} + F\cos(\omega t)$$
Which can be split into the system of differential equations:
$$\dot{x}(t) = y(t)$$
$$m\dot{y}(t) = -\nu y(t) + x(t) - x^3(t) + F\cos(\omega t)$$
Here, we will solve these using the Runge-Kutta 4th Order method outlined in the past two classworks, taking $m = 1$, $\omega = 1$ and $\nu = .25$. We'll graph our results and then see what everything means.

## Line and Scatter Plot Solutions
Below we have graphed the solutions calculated from the Runge-Kutta method of ODEs for different sets of initial conditions. Each is graphed parametrically, with each coordinate in the form $(x(t), y(t))$. The scatter plots take in to account specific $t$-values such that $t = n2\pi$.

### Graph 1:
Inital Conditions: $F = 0.18$, $t\in[0,2\pi\, 50]$, with $x(0) = -0.9$ and $y(0) = 0$

```python
plot = so.stingray(0.18, -0.9, 0, 50)
so.linegraph(plot, 0.18, "$F=0.18$, $x(0) = -0.9$, $y(0)= 0$")
```

```python
scatterpoints = np.array([plot[:,1000*(i+1)] for i in range(50)])
so.scattergraph(np.array([scatterpoints[:,0], scatterpoints[:,1]]), 0.18, "$F=0.18$, $x(0) = -0.9$, $y(0)= 0$")
```

Here, we can see that general path the ball follows when our stingray is thrown around by tides in his ocean habitat(I'm going to call him George because that's what Natanael would name him if I asked anyway). For very strong tides, we can see that George doesn't get to rest on one side of his mountain, but is instead swished around in a mostly uniform pattern. Luckily for him, the generally elliptical shape suggests he was resting on the ocean floor for most of the crazy tide episode. The scatter plot confirms this, as we see very few outlying points off the elliptical pattern.
